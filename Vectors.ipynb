```python
import fasttext
import numpy as np
import pandas as pd
```

```python
base_df = pd.read_csv(
    "../data/База по всем организациям и ИП.zip",
    sep=";", 
    chunksize=10e3, 
    dtype=str
)
```

```python
chunk = base_df.get_chunk(1000)
```

```python
names = chunk["Наименование полное"].str.upper().to_list()
names
```

```python
model = fasttext.load_model("/home/pavel/cc.ru.300.bin.gz")
```

```python
vectors = {
    name: np.array(model.get_sentence_vector(name))
    for name in names
}
```

```python
def search(name):
    vec = np.array(model.get_sentence_vector(name))
    res = []
    for name, vec2 in vectors.items():
        dist = np.linalg.norm(vec - vec2)
        res.append((name, dist))
        
    return sorted(res, key=lambda x: -x[1])[:5]
```

```python
test_name = ""
search(test_name)
```
